```
def myfunction(x):
    if x<-5:
       return 2*x+15 
    elif x<0:
      return -2*x-5
    elif x<5:
      return 2*x-5
    else:
      return -2*x+15
```

```
x=np.array(np.arange(-10,11,0.2)) #define range / another method => use radn function and adjust this into the target range.
pickness=100
```

```
observation=np.array(random.choices(x,k=pickness)) #randomly pick /infinite=> finite observation 
```

Synthesized function

```
ar=np.array([])
for a in observation:
  ar=np.append(ar,myfunction(a))
```

```
function_value=np.array([])
for a in x:
  function_value=np.append(function_value,myfunction(a))
```

Common function

```
ar=np.array([])
for a in observation:
  ar=np.append(ar,np.sin(a))
```

```
function_value=np.array([])
for a in x:
  function_value=np.append(function_value,np.sin(a))
```

Deep Neural Network

```
model=Sequential()
model.add(Dense(100, input_shape = (1,),activation='relu'))    # Input layer => input_shape should be explicitly designated  #input dimension = training data dimension
model.add(Dense(100,activation='relu'))                         # Hidden layer => only output dimension should be designated
model.add(Dense(1))                         # Output layer => output dimension = 1 since it is regression problem
```

```
sgd = optimizers.SGD(lr = 0.01) 
adam = optimizers.Adam(lr = 0.001)                     # use Adam optimizer
```

```
model.compile(optimizer = sgd, loss = 'mean_squared_error', metrics = ['mse'])
```

```
model.fit(observation,ar,batch_size=pickness, epochs=10000,verbose=1)
```

```
prediction_value=model.predict(x)
```

```
plt.plot(x,prediction_value)
```
